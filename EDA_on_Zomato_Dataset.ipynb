```python
df_country=pd.read_excel('/content/drive/MyDrive/Country-Code.xlsx')
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(5)
```

```python
#To check datatypes
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
print(country_names)
```

```python
country_values=final_df.Country.value_counts().values
print(country_values)
```

```python
#Pie Chart-Top 3
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation
Maximum number of trannsactions are from India,then  comes UK and then US


```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
print(ratings)
```

##Conclusions
1.Rating between 4.5 -4.9    Excellent
2.Rating between 4.0 -4.4    very Good
3.Rating between 3.5 -3.9     Good
4.Rating between 2.5 -3.4     Average
5.Rating between 1.8 -2.4     Poor
6.Rating          0           Not rated

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

Observation
1.Not rated count is very high
2.Maximun number of ratings are between 2.5 to 3.4

```python
#Count plot
sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green','green'])
```
