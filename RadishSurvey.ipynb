```python
#Create an empty dictionary for associating radish names with vote counts  counts = {}

counts = {}
voted = []

file= open("/content/drive/MyDrive/radishsurvey.txt","r")
for line in file:
    line = line.strip()
    personname, radishvariety = line.split(" - ")
    personname = personname.upper()
    if personname not in voted :
        voted.append(personname)
    else:
        print(personname,"has already voted" )
        continue
    radishvariety = radishvariety.strip().upper()
    radishvariety = radishvariety.replace("  "," ")
    if radishvariety not in counts:
            # First vote for this variety
        counts[radishvariety] = 1
    else:
            # Increment the vote count
        counts[radishvariety] = counts[radishvariety] + 1
print(counts)
print("-------------------------")

for key,value in counts.items():
    print(key,":",value)
print("-------------------------")
print("VOTED LIST")
for i in range(0,len(voted)-1):
    print(voted[i])
print("-------------------------")
print("WINNER")
winnervotes=0
winnernamre=" "
for radishvariety in counts:
    if counts[radishvariety]>winnervotes:
        winnervotes=counts[radishvariety]
        winnernamre =radishvariety
print("winnwer is :",winnernamre)
print("no of votes casted is : ",winnervotes)
```
